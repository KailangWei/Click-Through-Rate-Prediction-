```python
dtrain = xgb.DMatrix(data=XTr,label=YTr)
dval = xgb.DMatrix(data=XVal,label=YVal)

param = { 'objective': 'binary:logistic', 'max_depth':5,'min_child_weight':5}
# 'max_depth': 2, 'eta': 1,

num_round = 5000
evallist = [(dval, 'eval'), (dtrain, 'train')]
bst = xgb.train(param, dtrain, num_round, evallist,early_stopping_rounds=100)
```

```python
ypred = bst.predict(dval, ntree_limit=bst.best_ntree_limit)

def ll(YVal,ypred):
    ll = -np.mean(YVal*np.log(ypred)+(1-YVal)*np.log(1-ypred))
    return(ll)

ll(YVal,ypred)
```

```python
def lln(YVal,ypred):
    ll = np.mean(YVal*np.log(ypred)+(1-YVal)*np.log(1-ypred))
    return(ll)
```
